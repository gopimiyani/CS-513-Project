```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import OneHotEncoder, LabelEncoder,MinMaxScaler, StandardScaler, Imputer
from sklearn.model_selection import train_test_split, ParameterGrid
from sklearn.metrics import accuracy_score, confusion_matrix, mean_squared_error, log_loss
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
```

```python
Phones = pd.read_csv("C:\\Users\\Darp\\Documents\\GitHub\\CS-513-Project\\Final_Refined_Phone_Dataset.csv")
```

```python
Phones.info()
```

```python
numeric_columns = ['RAM', 'primary_camera', 'secondary_camera', 'weight_g','internal_memory', 'battery','Thickness', 'display_size', 'screen_to_body_ratio', 'Price']
categorical_columns = [c for c in Phones.columns if c not in numeric_columns]
print(categorical_columns)
print(numeric_columns)
```

```python
numeric = Phones[numeric_columns]
category = Phones[categorical_columns]
```

```python
num_imputer = SimpleImputer(missing_values=np.nan, strategy='mean')
numerical = pd.DataFrame(num_imputer.fit_transform(numeric))
numerical.columns = numeric.columns
#numerical = numeric.iloc[:,0:].fillna(numeric.iloc[:,0:].mean())
```

```python
scaler = MinMaxScaler()
scaler.fit(numerical)
Nor_numerical = pd.DataFrame(scaler.transform(numerical))
Nor_numerical.columns = numerical.columns
```

```python
Nor_numerical['Price'] = Phones['Price']
#Nor_numerical['Price'][5003] = Phones['Price'].mean()
```

```python
Nor_numerical.head(20)
```

```python
cat_imputer = SimpleImputer(missing_values=np.nan, strategy='most_frequent')
categorical = pd.DataFrame(cat_imputer.fit_transform(category))
categorical.columns = category.columns
#categorical = category.iloc[:,0:].fillna(category.iloc[:,0:].mode())
```

```python
le = LabelEncoder()
Cat = categorical.apply(le.fit_transform)
Cat.head()
```

```python
ohe = OneHotEncoder()
ohe.fit(Cat)
Catro = pd.DataFrame(ohe.transform(Cat).toarray())
Catro.shape
Catro.head()
```

```python
Phones_rna = pd.concat([Catro, Nor_numerical], axis=1, sort=False)
```

```python
Phones_rna.head(20)
```

```python
Phones_rna.to_csv("Final_Refined_OneHotEncoded.csv", index=False)
```
